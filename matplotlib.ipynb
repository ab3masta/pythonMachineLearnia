```python
import numpy as np
%matplotlib inline 
```

```python
x = np.linspace(0,2,10)
y = x**2
print(x)
```

```python
import matplotlib.pyplot as plt
```

```python
# x et y doivent avoir la meme dimension
# le paramettre c pour la couleur // lw pour l'epaisseur du trait // ls pour le style du trait comme --
plt.plot(x,y,c="red",lw=5,ls="--")
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

### Le cycle de vie d'une figure 
#### tres important on commen avec plt.figure() ==> plt.plot(x,y)
##### on peut changer la taille de la figure plt.figure(figsize(12 , 8))
##### en utilisant plt.figure() on peut ajouter des coubres sur la meme figure
##### pour le titre on utilise plt.title("figure name")
##### les nom pour les axes plt.xlabel("axe x") et plt.ylabel("axe y")
##### on peut aussi ajouter des legends donc pour les courbes on ajoute des labels ==> plt.plot(x,y, label='label')
##### et enfin plt.show()
##### pour avoir l'image ==> plt.savefig('text.png')

```python
plt.figure()
plt.plot(x,y,label="quadratique")
plt.plot(x,y**3,label="cubique")
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
plt.savefig('figure1.png')
```

Utiliser une grille de graphes

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,y,label="quadratique")
plt.plot(x,y**3,label="cubique")
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.savefig('figure1.png')
plt.subplot(2,1,2)
plt.plot(x,y,label="quadratique")
plt.plot(x,y**3,label="cubique")
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
```

### Scatter le graphique de classification

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x =iris.data
y = iris.target
names = list(iris.target_names)

print(f'x contient {x.shape[0]} examples et {x.shape[1]} variables')
print(f'il y a {np.unique(y).size} classes')
y
```

```python
# alpha ==> Controller la transparence de chaque point et s==> la taille des points avec s=100 ou controller la taille en fonction d'une des variables ==> s=x[:,2]*100
plt.scatter(x[:,0],x[:,1],c=y,alpha=0.5,s=x[:,2]*100)
plt.xlabel('longeur sepal')
plt.ylabel('largeur sepal')
```

```python
f = lambda x,y: np.sin(x) + np.cos(x+y)
```

```python
X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)
X,Y = np.meshgrid(X,Y)

Z = f(X,Y)
Z.shape
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='plasma')
```

```python
dataset = {f"experience {i}":np.random.randn(100,3) for i in range(4)}
```

```python
# code tres utile
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12,8))
    for k,i in zip(data.keys() , range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```

```python
graphique(dataset)
```
